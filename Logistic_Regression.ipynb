```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=10,n_features=2,n_redundant=0,n_informative=2,random_state=1,class_sep=30)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    w=np.ones(X.shape[1])
    lr=0.1

    for i in range(100):
       j=np.random.randint(0,X.shape[0])
       y_hat=step(np.dot(w,X[j]))
       w=w+lr*(y[j]-y_hat)*X[j]

    return w

def step(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
pr=perceptron(X,y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(X[:,0],-(pr[0]+pr[1]*X[:,0])/pr[2])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr=LogisticRegression()
logr.fit(X,y)
m=-(logr.coef_[0][0]/logr.coef_[0][1])
c=-(logr.intercept_[0]/logr.coef_[0][1])
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(X[:,0],-(pr[0]+pr[1]*X[:,0])/pr[2],color='red',label="Perceptron")
plt.plot(X[:,0],m*X[:,0]+c,color='green',label='Regression Line')
plt.legend()
plt.show()
```
