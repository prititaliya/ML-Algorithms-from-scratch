```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class LinearRegressionScratch:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        self.m=None
        self.b=None

    def fit(self, X, y):
       num=0
       den=0
       for i in range(X.shape[0]):
           num+=(X[i]-np.mean(X))*(y[i]-np.mean(y))
           den+=(X[i]-np.mean(X))**2
       self.m=num/den
       self.b=np.mean(y)-(self.m*np.mean(X))
       print(self.m,self.b)


    def predict(self, X):
        y_pred = self.m * X + self.b
        return y_pred
```

```python
scr_lin=LinearRegressionScratch()
scr_lin.fit(X_train,y_train)
y_preds=scr_lin.predict(X_test)
mean_squared_error(y_test,y_preds)
```

```python
class multipleLinearRegression:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.m=None
        self.b=None

    def fit(self,X,y):
        X=np.insert(X,0,1,axis=1)
        self.m=np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))
        self.b=self.m[0]
        self.m=self.m[1:]

    def predict(self,X):
        y_pred=np.dot(X,self.m)+self.b
        return y_pred
```

```python
scr_lin=multipleLinearRegression()
scr_lin.fit(X_train,y_train)
y_preds=scr_lin.predict(X_test)
mean_squared_error(y_test,y_preds)
```

```python
class GradientD_regressor:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.m=0
        self.b=0

    def fit(self, X, y):
        for i in range(self.n_iterations):
            l_slope_b=-2*np.sum(y-self.m*X.ravel()-self.b)
            l_slope_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())

            self.b=self.b-(self.learning_rate*l_slope_b)
            self.m=self.m-(self.learning_rate*l_slope_m)

    def predict(self, X):
        return self.m*X.ravel()+self.b
```
